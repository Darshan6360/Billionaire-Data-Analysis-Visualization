```python
from matplotlib import pyplot as p
import numpy as np
```

```python
x=np.array([10,20])
y=np.array([20,40])
```

```python
p.plot(x,y)
```

```python
p.plot(x,y,"o")
```

```python
p.plot(x,y,marker="*")
```

```python
p.plot(x,y,'p--r')
#after the - it takes color of the line r=red, b=blue, k=black
#for line - plane line
```

```python
p.plot(x,y,'p--r',ms=10)
#markersize ms
```

```python
p.plot(x,y,"v--r",ms=10,mfc='k')
#markerface color mfc
#the kind of marker p,o,"v",etc
```
